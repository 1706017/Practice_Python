```python
#Importing matplotlib library

import matplotlib.pyplot as plt
```

```python
#To enable the plot so that they can be visible in notebook
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.arange(1,21)
```

```python
y=x**2
```

```python
x
```

```python
y
```

```python
#Function approach to draw plot 
plt.plot(x,y)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("X VS Y")

```

```python
plt.subplots(1,2)
plt.plot(x,y)
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("X vs Y ")



```

```python
#Object Oriented Approach for plotings

fig=plt.figure()
axes=fig.add_axes([0,0,1,1]) #Here the flow is left,bottom,right,top
axes.plot(x,y,color='red',ls='--',lw='2') #here lw=line width,ls=line style
```

```python
fig=plt.figure()
axes1=fig.add_axes([0,0,1,1])
axes1.plot(x,y,color="Yellow",label="X vs Y")
axes1.legend()
axes1.set_xlabel("X label")
axes1.set_ylabel("Y label")
axes1.set_title("X VS Y")

axes2=fig.add_axes([0.5,0.5,0.1,0.2])
axes2.plot(y,x,color='Red',label='Y VS X')
axes2.legend()
axes2.set_xlabel("Xlabel")
axes2.set_ylabel("Y label")
axes2.set_title("Y vs X")
```
